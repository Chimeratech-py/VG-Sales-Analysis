# Data Cleaning
To attain an effective data analysis, we'll first have to handle missing data, as well as converting any columns as necessary.

**Type conversion**

First, let's find out why Year originally has its data type as float. Let's see any anomalous values:

```python
#Check decimals
df_has_decimal = vgsales[(vgsales['Year']%1 != 0) & vgsales['Year'].notnull()]
df_has_decimal
```

```python
#Check null Years
null_years = vgsales[vgsales['Year'].isnull()]
null_years
```

From here, we can tell that there are no decimal values for the year, but there are indeed null values, as shown by NaN.

We also spotted rows that have Publisher as "Unknown" instead of NaN.

We'll solve these problems step by step



```python
#replace values with "Unknown" to null so they can be processed correctly
vgsales['Publisher'] = vgsales['Publisher'].replace('Unknown', np.nan)
```

Now it would be easier to see which rows won't be very helpful to our study

```python
#Get all rows with either null years or null publishers
all_nulls = vgsales[vgsales['Year'].isnull() | vgsales['Publisher'].isnull()]
all_nulls
```

Now we'll try to find out how much these rows contribute to the sum of the global sales. <br>
That way, we can see if we can safely drop those rows.

```python
#sum of the global sales of rows with missing years
sum_of_missing = all_nulls['Global_Sales'].sum()
print(sum_of_missing)
```

```python
#Find the sum of the Global_sales
sum_of_global_sales = vgsales['Global_Sales'].sum()
print(sum_of_global_sales)
```

```python
#Find the percentage to see how much the games with missing years contribute in terms of global sales
percent_missing = (sum_of_missing/sum_of_global_sales) * 100
print(percent_missing)
```

We've determined that the rows with missing data only contribute 1.37% of the global sales. As such, we can conclude that we're not missing much, so it's safe to drop those rows.

```python
#create a new dataframe without nulls
vgsales_notnull = vgsales[vgsales['Year'].notnull() & vgsales['Publisher'].notnull()]
vgsales_notnull
```

Everything's consistent with what we expect. <br> 
There are 407 rows with missing values, and we removed that from the original 16598 rows, so now we're left with 16191.<br><br>

Now, let's remove duplicates

```python
# Find the number of duplicate data
print(vgsales_notnull.duplicated().sum())
```

Great! We don't have duplicated data. we can now move on

# Exploratory Data Analysis

## 1. Which among our competitors have games that have the highest sales?

```python
competitor_analysis = pd.DataFrame(vgsales_notnull.groupby(['Publisher'])[['Global_Sales']].sum())
competitor_analysis.sort_values(by='Global_Sales', ascending=False, inplace=True)
competitor_analysis_10 = competitor_analysis.head(10)
competitor_analysis_10
```

```python
# Set for grouped plots - figure with a 2x2 grid of Axes
sns.set_style('darkgrid')
sns.barplot(x='Global_Sales', y='Publisher', data = competitor_analysis_10)
plt.show()
```

## 2. Which among our competitors have consistently made best-selling video games?

```python
top_games = (
    vgsales_notnull
    .groupby(['Name', 'Publisher'], as_index=False)['Global_Sales']
    .sum()
    .sort_values(by='Global_Sales', ascending=False)
    .head(100)
)
```

```python
top_games
```

```python
consistent_publishers = (
    top_games
    .groupby('Publisher', as_index=False)  # or 'publisher_name' if available
    .agg(game_count=('Name', 'count'))
)

# 4. Filter publishers with > 5 top-selling games
consistent_publishers = consistent_publishers[consistent_publishers['game_count'] > 5]

# 5. Sort by game count
consistent_publishers = consistent_publishers.sort_values(by='game_count', ascending=False)
consistent_publishers
```

```python
sns.set_style('darkgrid')
sns.barplot(x='game_count', y='Publisher', data = consistent_publishers)
plt.show()
```

## 3. Which genres sell the best per region?

```python
NA_best_genre = (
    vgsales_notnull
    .groupby(['Genre'], as_index=False)['NA_Sales']
    .sum()
    .sort_values(by='NA_Sales', ascending=False)
    .head(100)
)
NA_best_genre
```

```python
sns.set_style('darkgrid')
sns.barplot(x='NA_Sales', y='Genre', data = NA_best_genre)
plt.show()
```

```python
JP_best_genre = (
    vgsales_notnull
    .groupby(['Genre'], as_index=False)['JP_Sales']
    .sum()
    .sort_values(by='JP_Sales', ascending=False)
    .head(100)
)
JP_best_genre
```

```python
sns.set_style('darkgrid')
sns.barplot(x='JP_Sales', y='Genre', data = JP_best_genre)
plt.show()
```

```python
EU_best_genre = (
    vgsales_notnull
    .groupby(['Genre'], as_index=False)['EU_Sales']
    .sum()
    .sort_values(by='EU_Sales', ascending=False)
    .head(100)
)
EU_best_genre
```

```python
sns.set_style('darkgrid')
sns.barplot(x='EU_Sales', y='Genre', data = EU_best_genre)
plt.show()
```

```python
best_platforms = (
    vgsales_notnull
    .groupby(['Platform'], as_index=False)['Global_Sales']
    .sum()
    .sort_values(by='Global_Sales', ascending=False)
    .head(15)
)
best_platforms
```

```python
sns.set_style('darkgrid')
sns.barplot(x='Global_Sales', y='Platform', data = best_platforms)
plt.show()
```
